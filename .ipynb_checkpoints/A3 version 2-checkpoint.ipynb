```python
import pandas as pd
import altair as alt
import numpy as np
alt.renderers.enable('notebook')
alt.data_transformers.enable('default', max_rows=None)
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_json('happiness_pivot.json')
df.loc[df['Region']=="Australia and New Zealand",'Region'] = 'North America, Aus, NZ'
df.loc[df['Region']=="North America",'Region'] = 'North America, Aus, NZ'
```

```python
region_scale = alt.Scale(range=['#DBD56E', '#88AB75', '#2D93AD', '#7D7C84', '#DE8F6E','#99D5C9','#33658A'])
measure_scale = alt.Scale(range=['#DBD56E', '#88AB75', '#2D93AD', '#7D7C84', '#DE8F6E','#99D5C9','#33658A'])

region_color = alt.Color('Region:N', scale=region_scale)
measure_color = alt.Color('happiness_measure:N', scale=measure_scale)

year_slider = alt.binding_range(min=2006, max=2017, step=1)
year_selector = alt.selection_single(bind=year_slider, fields=['Year'], name="Year")

region_selector = alt.selection_single(empty='all', fields=['Region'])
country_selector = alt.selection_single(empty='all', fields=['Country'])


base = alt.Chart(df).properties(
    width=250,
    height=250
)

gdp_region = base.mark_point(filled=True, size=200).encode(
    x=alt.X('mean(Log GDP per capita)', scale=alt.Scale(domain=(6,12))),
    y=alt.Y('mean(Life Ladder)', scale=alt.Scale(domain=(0, 10))),
    color=alt.condition(region_selector, region_color, alt.value('lightgray')),
).transform_filter(
    year_selector
).add_selection(
    year_selector
).add_selection(
    region_selector
).interactive()

gdp_country = base.mark_circle().encode(
    x=alt.X('mean(Log GDP per capita)', scale=alt.Scale(domain=(6,12))),
    y=alt.Y('mean(Life Ladder)', scale=alt.Scale(domain=(0, 10))),
    color=alt.Color('Region:O'),
    tooltip=['Country:N']
).transform_filter(
    region_selector
).transform_filter(
    year_selector
).transform_filter(
    country_selector
).add_selection(
    year_selector
).add_selection(
    country_selector
)

life_region = base.mark_point(filled=True, size=200).encode(
    x=alt.X('mean(Healthy life expectancy at birth)', scale=alt.Scale(domain=(30,80))),
    y=alt.Y('mean(Life Ladder)', scale=alt.Scale(domain=(0, 10))),
    color=alt.condition(region_selector, region_color, alt.value('lightgray')),
).transform_filter(
    year_selector
).add_selection(
    year_selector
).add_selection(
    region_selector
).interactive()

life_country = base.mark_circle().encode(
    x=alt.X('mean(Healthy life expectancy at birth)', scale=alt.Scale(domain=(30,80))),
    y=alt.Y('mean(Life Ladder)', scale=alt.Scale(domain=(0, 10))),
    color=alt.Color('Region:O'),
    tooltip=['Country:N']
).transform_filter(
    region_selector
).transform_filter(
    year_selector
).transform_filter(
    country_selector
).add_selection(
    year_selector
).add_selection(
    country_selector
)

# Bottom panel is a bar chart of weather type
bars = alt.Chart(df).mark_bar().encode(
    y = alt.Y('happiness_measure:N', axis = alt.Axis(title=None, labelAngle=0)),
    x = alt.X('binary_val:Q', 
              axis = alt.Axis(format='%',title=None), 
              scale = alt.Scale(domain=(-0.32,1))
             ),
    color=alt.value('gray'),
).transform_filter(
    year_selector
).transform_filter(
    region_selector
).transform_filter(
    country_selector
)

((gdp_region | gdp_country) & (life_region | life_country)) & bars
```
