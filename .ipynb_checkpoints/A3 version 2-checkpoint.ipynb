```python
import pandas as pd
import altair as alt
import numpy as np
alt.renderers.enable('notebook')
#alt.data_transformers.enable('default', max_rows=None)
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_json('happiness_big.json')
```

```python
df.loc[df['region']=="Australia and New Zealand",'region'] = 'North America, Aus, NZ'
df.loc[df['region']=="North America",'region'] = 'North America, Aus, NZ'
df
```

```python
# A slider filter


brush = alt.selection(type='interval', encodings=['x','y'])
brush = alt.selection_single(empty='all')

year_slider = alt.binding_range(min=2006, max=2017, step=1)
slider_selection = alt.selection_single(bind=year_slider, fields=['year'], name="Year")

selector = alt.selection_single(empty='all', fields=['region'])


base = alt.Chart(df).mark_circle(size=10).encode(
    alt.Y('happy:Q', axis = alt.Axis(title='Happiness Score')),
    color='region:N',
    opacity=alt.condition(brush, alt.OpacityValue(1), alt.OpacityValue(0.1)),
    tooltip=['country:N']
).add_selection(
    brush
).add_selection(
    slider_selection
).transform_filter(
    slider_selection
).properties(
  width=240,
  height=180
).transform_filter(
    'datum.year!=null'
).interactive()


up_left = base.encode(x='gdp_per_cap').properties(title="GDP per Capita").add_selection(
    brush
)
up_right = base.encode(x='life_expect').properties(title="Life Expectancy").add_selection(
    brush
)
bottom_left = up_left.transform_filter(
selector
)
bottom_right = up_right.transform_filter(
selector
)


bottom_left | bottom_right
#alt.vconcat(
#    alt.hconcat(up_left,up_right),
#    alt.hconcat(bottom_left,bottom_right)
#)

```

```python
from vega_datasets import data

source = data.seattle_weather()
brush = alt.selection(type='interval', encodings=['x'])

bars = alt.Chart().mark_bar().encode(
    x='month(date):O',
    y='mean(precipitation):Q',
    opacity=alt.condition(brush, alt.OpacityValue(1), alt.OpacityValue(0.7)),
).add_selection(
    brush
)

line = alt.Chart().mark_rule(color='firebrick').encode(
    y='mean(precipitation):Q',
    size=alt.SizeValue(3)
).transform_filter(
    brush
)

alt.layer(bars, line, data=source)
```

```python

np.random.seed(0)

n_objects = 20
n_times = 50

# Create one (x, y) pair of metadata per object
locations = pd.DataFrame({
    'id': range(n_objects),
    'x': np.random.randn(n_objects),
    'y': np.random.randn(n_objects)
})

# Create a 50-element time-series for each object
timeseries = pd.DataFrame(np.random.randn(n_times, n_objects).cumsum(0),
                          columns=locations['id'],
                          index=pd.RangeIndex(0, n_times, name='time'))

# Melt the wide-form timeseries into a long-form view
timeseries = timeseries.reset_index().melt('time')

# Merge the (x, y) metadata into the long-form view
timeseries['id'] = timeseries['id'].astype(int)  # make merge not complain
data = pd.merge(timeseries, locations, on='id')

# Data is prepared, now make a chart

selector = alt.selection_single(empty='all', fields=['id'])

base = alt.Chart(data).properties(
    width=250,
    height=250
).add_selection(selector)

points = base.mark_point(filled=True, size=200).encode(
    x='mean(x)',
    y='mean(y)',
    color=alt.condition(selector, 'id:O', alt.value('lightgray'), legend=None),
).interactive()

timeseries = base.mark_line().encode(
    x='time',
    y=alt.Y('value', scale=alt.Scale(domain=(-15, 15))),
    color=alt.Color('id:O', legend=None)
).transform_filter(
    selector
)

points | timeseries
```

```python
colorscale = alt.Scale(range=['#DBD56E', '#88AB75', '#2D93AD', '#7D7C84', '#DE8F6E','#99D5C9','#33658A'])
color = alt.Color('region:N', scale=colorscale)


year_slider = alt.binding_range(min=2006, max=2017, step=1)
year_selector = alt.selection_single(bind=year_slider, fields=['year'], name="Year")

region_selector = alt.selection_single(empty='all', fields=['region'])
country_selector = alt.selection_single(empty='all', fields=['country'])


base = alt.Chart(df).properties(
    width=250,
    height=250
)

gdp_region = base.mark_point(filled=True, size=200).encode(
    x=alt.X('mean(gdp_per_cap)', scale=alt.Scale(domain=(6,12))),
    y=alt.Y('mean(happy)', scale=alt.Scale(domain=(0, 10))),
    color=alt.condition(region_selector, color, alt.value('lightgray')),
).transform_filter(
    year_selector
).add_selection(
    year_selector
).add_selection(
    region_selector
).interactive()

gdp_country = base.mark_circle().encode(
    x=alt.X('gdp_per_cap', scale=alt.Scale(domain=(6,12))),
    y=alt.Y('happy', scale=alt.Scale(domain=(0, 10))),
    color=alt.Color('region:O'),
    tooltip=['country:N']
).transform_filter(
    region_selector
).transform_filter(
    year_selector
).transform_filter(
    country_selector
).add_selection(
    year_selector
).add_selection(
    country_selector
)

life_region = base.mark_point(filled=True, size=200).encode(
    x=alt.X('mean(life_expect)', scale=alt.Scale(domain=(30,80))),
    y=alt.Y('mean(happy)', scale=alt.Scale(domain=(0, 10))),
    color=alt.condition(region_selector, color, alt.value('lightgray')),
).transform_filter(
    year_selector
).add_selection(
    year_selector
).add_selection(
    region_selector
).interactive()

life_country = base.mark_circle().encode(
    x=alt.X('life_expect', scale=alt.Scale(domain=(30,80))),
    y=alt.Y('happy', scale=alt.Scale(domain=(0, 10))),
    color=alt.Color('region:O'),
    tooltip=['country:N']
).transform_filter(
    region_selector
).transform_filter(
    year_selector
).transform_filter(
    country_selector
).add_selection(
    year_selector
).add_selection(
    country_selector
)

alt.vconcat(
    gdp_region | gdp_country,
    life_region | life_country
)
```
